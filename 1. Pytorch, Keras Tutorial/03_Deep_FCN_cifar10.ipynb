```python
import torch
import torchvision #이미지를 다루기 위한 파이토치
import torch.nn as nn ##neural net을 활용하기 위한 파이토치
import numpy as np
import torchvision.transforms as transforms # 데이터 shuffle과 augementation을 위함
import matplotlib.pyplot as plt
```


####  cifar10 데이터셋

총 10개의 클래스를 가지는 이미지와 라벨 데이터를 가지고 있는 데이터셋
32 x 32 사이즈의 크기이며, 컬러를 가짐.
train data 50000개 test data 10000개로 구성


```python
### CIFAR 10 DataSet Loading and Show Image
```

```python
train_dataset = torchvision.datasets.CIFAR10(root = '../../data/', # 데이터셋이 저장되는 물리적 위치
                                             train = True, # Train dataset만 저장, False면 test dataset만 저장
                                             transform = transforms.ToTensor(), # 중요!! 0~255사이의 픽셀값들을 스케일링을 처리하여 0~1사이의 값으로 치환해준다... 채널을 앞으로 보낸다. 나중에 다시 뒤로 보내야함.
                                             download = True # 다운로드가 root에 진행됨
                                            )
```

```python
## 다운로드된 이미지 실제로 찍어볼 수 있다.
print(train_dataset[0]) # 이미지 스케일링된 픽셀값과 ,를 기준으로 인덱스값 출력
image, label = train_dataset[0]

print('image 픽셀 값 : ', image)
print('label : ', label)

print('Image size : ' , image.size()) # 32 x 32 x 3 --> [chanel, height, width]로 출력
```

#### Image 변환하기
    Torch타입을 Numpy로 일단 변경부터 시키기. 채널이 앞으로 나와있으므로

```python
image = image.numpy() # 이 과정으로부터 3x32x32 --> 32 x 32 x 3 으로 변경해줘야함
image = np.moveaxis(image, 0, -1)
print(image.shape)

plt.imshow(image)
```

### CIFAR 10 DataSet Loading Using BatchSize

```python
train_loader = torch.utils.data.DataLoader(dataset=train_dataset,
                                           batch_size=64,
                                           shuffle=True)

# batch_size는 전체 데이터를 batch_size크기로 쪼개서 여러번 나누어 데이터를 로딩
# 셔플은 batch_size로 나뉜 데이터를 섞는다.학습의 효율성을 높이는데 기여.
# 데이터를 저장하고 load하는 데에는 이러한 과정들이 전부 포함된다.

cnt = 0
for images, labels in train_loader:
    '''
    이 안에서 실질적으로 어떤 작업이 진행되는가?
    model 생성
    loss, optimizer 선 정의
    model 값 입력.. 예측값 반환
    loss 값
    backward()
    학습 진행 step()
    '''
    cnt = cnt+1
print('cnt : ', cnt) #782 .. drop_last 속성 기본값이 False로 잡혀있으므로
```
